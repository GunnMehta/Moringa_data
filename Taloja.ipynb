```python
# @title Date vs NO2

from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['Date']
  ys = series['NO2']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = df.sort_values('Date', ascending=True)
for i, (series_name, series) in enumerate(df_sorted.groupby('Time')):
  _plot_series(series, series_name, i)
  fig.legend(title='Time', bbox_to_anchor=(1, 1), loc='upper left')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Date')
_ = plt.ylabel('NO2')
```

```python
import pandas as pd
import plotly.graph_objects as go

# Load the dataset
file_path = "/content/comparisionHourly_Taloja- till 5th Mar-V2.csv"  # Update with correct path
df = pd.read_csv(file_path)

# Clean column names (remove extra spaces)
df.columns = df.columns.str.strip()

# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], format="%d-%m-%Y")

# Group data by Date and calculate pollution statistics
pollutant = "PM2.5"  # Change this for other pollutants
df_grouped = df.groupby("Date")[pollutant].agg(["min", "max", "first", "last"]).reset_index()

# Create Upstock (Candlestick-style) chart
fig = go.Figure(data=[go.Candlestick(
    x=df_grouped["Date"],
    open=df_grouped["min"],  # Open = Min pollution of the day
    high=df_grouped["max"],  # High = Max pollution of the day
    low=df_grouped["min"],   # Low = Min pollution of the day
    close=df_grouped["last"], # Close = Last pollution level of the day
    name=pollutant
)])

# Chart formatting
fig.update_layout(
    title=f"{pollutant} Levels Over Time (Upstock Style)",
    xaxis_title="Date",
    yaxis_title=f"{pollutant} Level",
    xaxis_rangeslider_visible=False
)

# Show plot
fig.show()
```
